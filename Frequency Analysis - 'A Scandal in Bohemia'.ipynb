### We'll now create a dictionary `alphadict` which holds the data about frequency of each letter in the story, and another variable `alphacount` which holds the total number of letters in the whole story.

```python
alphabets = "abcdefghijklmnopqrstuvwxyz"
alphadict = dict()
alphacount = 0

for alphabet in alphabets:
    alphadict[alphabet] = 0

for alphabet in page:
    if alphabet in alphabets:
        alphadict[alphabet] += 1
        alphacount += 1

print(f'Frequency of alphabets:\n{alphadict}\n')
print('Total number of alphabets:',alphacount)
```

### Now we're going to create a DataFrame for the frequency % of letters we have gathered from the story. We're also going to add another column to this DataFrame which holds the relative frequency % of every alphabet in the English language.
[Data for relative frequency % obtained from Wikipedia](https://en.wikipedia.org/wiki/Letter_frequency)

```python
df = pd.DataFrame.from_dict(alphadict, orient='index').rename(columns={0:'Observed Frequency (%)'})
df['Observed Frequency (%)'] = (df['Observed Frequency (%)']/alphacount)*100
relative_freq = [8.497,1.492,2.202,4.253,11.162,2.228,2.015,6.094,7.546,0.153,1.292,4.025,2.406,6.749,7.507,1.929,0.095,7.587,6.327,9.356,2.758,0.978,2.560,0.150,1.994,0.077]
df['Relative Frequency (%)'] = relative_freq
df
```

### Frequency of each letter in the story in decreasing order

```python
for i in sorted(alphadict.items(), key=lambda k:k[1], reverse=True):
    print(f'{i[0]} : {i[1]}')
```

### Percentage error of our observed frequency % vs. relative frequency % from Wikipedia

```python
percent_diff = dict()
for i in range(26):
    percent_diff[alphabets[i]]=(abs(df['Observed Frequency (%)'][i]-relative_freq[i])/relative_freq[i])*100

print('% error in observed frequency % vs. relative frequency % (rounded to 2 decimal places):\n')

for i in sorted(percent_diff.items(), key=lambda k:k[1]):
    print(f'{i[0]} : {round(i[1],2)} %')
```
